```python
import json, time, re
import requests
import trafilatura
from bs4 import BeautifulSoup
from datetime import datetime

HEADERS = {
    "User-Agent": "dt-support-copilot-poc/0.1 (educational; contact: internal)"
}

def extract_title(html: str) -> str:
    soup = BeautifulSoup(html, "html.parser")
    if soup.title and soup.title.text:
        return re.sub(r"\s+", " ", soup.title.text).strip()
    h1 = soup.find("h1")
    if h1 and h1.text:
        return re.sub(r"\s+", " ", h1.text).strip()
    return "Untitled"

def fetch_and_extract(url: str, topic: str, source_type: str = "docs") -> dict:
    r = requests.get(url, headers=HEADERS, timeout=30)
    r.raise_for_status()
    html = r.text

    title = extract_title(html)

    text = trafilatura.extract(
        html,
        include_comments=False,
        include_tables=True,
        include_formatting=True
    )

    if not text or len(text.strip()) < 200:
        raise ValueError("Extraction too short / empty")

    return {
        "url": url,
        "title": title,
        "topic": topic,
        "source_type": source_type,
        "fetched_at": datetime.utcnow().isoformat() + "Z",
        "text": text.strip()
    }
```

```python
OUT_PATH = Path("../data/raw/docs.jsonl")

records = []
errors = []

for topic, urls in SEED_URLS.items():


    for url in urls:
        try:
            source_type = "community" if topic == "community" else "docs"
            rec = fetch_and_extract(url=url, topic=topic, source_type=source_type)
            records.append(rec)
            print("✅", topic, "-", rec["title"])
            time.sleep(1.0)  # polite delay
        except Exception as e:
            errors.append({"url": url, "topic": topic, "error": str(e)})
            print("❌", topic, "-", url, "->", e)

# write JSONL
with OUT_PATH.open("w", encoding="utf-8") as f:
    for rec in records:
        f.write(json.dumps(rec, ensure_ascii=False) + "\n")

print("\nSaved:", OUT_PATH, "records:", len(records), "errors:", len(errors))
```

```python
# show first 1–2 records
for rec in records[:5]:
    print("\n====================")
    print("Title:", rec["title"])
    print("Topic:", rec["topic"])
    print("URL:", rec["url"])
    print("Snippet:", rec["text"][:1500], "...")
```
